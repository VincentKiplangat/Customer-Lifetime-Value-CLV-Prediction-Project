```python
# importing necessaries libraries
#basic
import pandas as pd

#visualization
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
case = pd.read_csv("customer_cases.csv")
case.head()
```

```python
case.info()
```

```python
case = case.drop(columns=["Unnamed: 0"])
case.head()
```

```python
info = pd.read_csv("customer_info.csv")
info.head()
```

```python
info = info.drop(columns=["Unnamed: 0"])
info.head()
```

```python
info.info()
```

```python
merge_one = pd.merge(case, info, on="customer_id")
merge_one.head()
```

```python
product = pd.read_csv("customer_product.csv")
product.head()
```

```python
product = product.drop(columns=["Unnamed: 0"])
product.head()
```

```python
merge_two = pd.merge(merge_one, product, on= "customer_id")
merge_two.head()
```

```python
p_info = pd.read_csv("product_info.csv")
p_info.head()
```

```python
p_info.rename(columns={"product_id": "product"}, inplace=True)
p_info.head()
```

```python
data = pd.merge(merge_two, p_info, on="product")
data.head()
```
